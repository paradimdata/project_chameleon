```python
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
tree = ET.parse('D:/Chameleon_backup/XRD/Bruker brml/practice/experiment0/RawData0.xml')
root = tree.getroot()
```

```python
root[1].text
```

```python
string_array = []
```

```python
for Datum in root.iter('Datum'):
    print(Datum.text)
    string_array.append(Datum.text)
```

```python
print("Root is",root)
```

```python
string_array[0]
```

```python
output_2d_array = []
```

```python
for string_element in string_array:
    # Split the string into a list of values using commas
    values = string_element.split(',')
    values = [float(val) for val in values]
    output_2d_array.append(values)
```

```python
output_2d_array[0][0]
```

```python
plt.plot(output_2d_array[2],output_2d_array[3], marker='o', linestyle='-', color='b', label='Data Points')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Graph of Two Arrays')
plt.show()
```
