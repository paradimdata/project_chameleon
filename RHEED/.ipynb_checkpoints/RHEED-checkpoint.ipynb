```python
import numpy as np 
import matplotlib.pyplot as plt
with open("YM16_LAO[110]_1.img","r") as f:
    f.seek(685)
    laue = np.fromfile(f,dtype="<u2",count=480*640).reshape((480,640))
print(laue)
im_temp = laue - laue.min()
im_temp = im_temp / im_temp.max()
im_temp = im_temp * 255

im_uint8_scaled = im_temp.astype(np.uint8)

plt.imshow(im_uint8_scaled)
plt.show()
```
